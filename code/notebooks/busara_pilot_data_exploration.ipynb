```python
import pandas as pd
from os.path import join
from glob import glob
from matplotlib import pyplot as plt

%matplotlib notebook

data_dir = "/Users/ianbolliger/git/berkeley_temp_behavior/data/raw/busara/pilot_temp_data/"
cool_data = join(data_dir,"20170915_Room2_Cool")
hot_data = join(data_dir,"20170915_Room1_Warm")
```

```python

fig,ax = plt.subplots(1,2,figsize=(8,4))
for ix,d in enumerate([cool_data,hot_data]):
    all_temps = []
    files = glob(join(d,'*.csv'))
    for f in files:
        station = f[-6:-4]
        df = pd.read_csv(f, skiprows=1,index_col = 1, parse_dates = True)
        temp = df.iloc[:,1]
        temp = temp[temp.notnull()]
        temp.name = 'temp_{}'.format(station)
        
        all_temps.append(temp)
    for t in all_temps:
        t.plot(ax=ax[ix])
```

```python
for i in all_temps:
    i.plot()
```
